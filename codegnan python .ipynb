```python
# set operations: union,intersection,difference,symmetric difference
s1 = {1,2,3,4,5}
s2 = {5,6,7,8,9}
# for union: (|,union())
print(s1 | s2) # dispalys all the elements in both sets 1 and 2
print(s1.union(s2))
# for  intersection:(&,intersection())
print(s1 & s2)
print(s1.intersection(s2))
# for difference : (-,difference())
print(s1 - s2)
print(s1.difference(s2))
# for symmetric difference : (^,symmetric_difference)
print(s1 ^ s2)
print(s1.symmetric_difference(s2))
# for subset: (<= or issubset())
print(s1 <= s2)
print(s1.issubset(s2))
# for superset: (>= or issuperset())
print(s1 >= s2)
print(s1.issuperset(s2))
```

```python
'''Sum of Digits Until Single Digit
Write a program to repeatedly add the digits of a number until the result is a single digit.
Example: 9875  9+8+7+5=29  2+9=11  1+1=2.'''
n = int(input('number:'))
while n >= 10:
  total = 0
  while n > 0:
    last = n % 10
    total += last
    n = n // 10
  n = total
print(n)

```

```python
'''2. Rotate Digits of a Number
Given a number, rotate its digits by k places.
 Example: Input: 12345, k=2  Output: 34512.
'''
n = int(input('number:'))
while n >= 10:
  n = 0
  temp = n
  while temp > 0:
    n+= temp
    temp //= 10
  print(n)
```

```python
# Sum of Digits Until Single Digit
#Write a program to repeatedly add the digits of a number until the result is a single digit.
#Example: 9875  9+8+7+5=29  2+9=11  1+1=2.
n = int(input('number:'))
while n >= 10:
  total = 0
  while n > 0:
    last = n % 10
    total += last
    n = n // 10
  n = total
  print(n)
```

**FUNCTIONS**
function -- resuable block of code to do a specific task,
def is the keyword to define a function
syntax:
def fun_name(parameter1,para2,......):
    # function body with indentation
    return value # (value out of the code)
function call -- will not run with definition, to run that block of code we have to make function call
function call -- has to be made using function name by passing values to the pareameters.these actual values are arguments
ex: fun_name(arg1,arg2,....)
key points :
> def is keyword to define a function
> parameters are variable used to pass the vaklues to the function
> return keyword is used to get values out of the function
> if we do not write any return statement,by default function returns none
> once the control reaches return, it will be the last statementsanythong written return statement will not be excuted.




```python
# waf to print product of two numbers (without return method)
def product(a,b): # a,b are parameters
  pr = a*b
  print(pr)
print("Start")
product(5,4) # here 5,4 are arguments
print("End")
product(8,5)
product(6,7)
```

**generator**

```python
## wagf that can produce infinite series of natural numbers
def fun_gen():
  i = 1
  while True:
    yield i
    i += 1
new = fun_gen()
print(next(new))
for i in range(5):
  print(next(new))
```

```python
# wagf that can produce factorial number
def fact_gen(n):
  i = 1
  while True:
    if i == 0:
      yield i
    else:
      yield n ** i
    i += 1
new = fact_gen(2)
print(next(new))
for i in range(5):
  print(next(new))
```
